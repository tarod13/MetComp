```python
divisores_a = [];
i = a
while i > 0:
    if (a % i) == 0:
        divisores_a.append(i)
    i = i-1

print "Los divisores de", a, "son", divisores_a
```

```python
divisores_b = [];
i = b
mcd = 1
while i > 0:
    if (b % i) == 0:
        divisores_b.append(i)
    i = i-1

if a > b:
    for div_b in divisores_b:
        if div_b in divisores_a:
            mcd = div_b
            break
else: 
    for div_a in divisores_a:
        if div_a in divisores_b:
            mcd = div_a
            break
    
print "El maximo comun divisor entre", a, "y", b, "es", mcd
```

```python
# define the Vehicle class
class Vehicle:
    name = ""
    kind = "car"
    color = ""
    value = 100.00
    def description(self):
        desc_str = "%s is a %s %s worth $%.2f." % (self.name, self.color, self.kind, self.value)
        return desc_str
    
# your code goes here
car1 = Vehicle()
car1.color = "red"
car1.kind = "convertible"
car1.value = 60000.00
car1.name = "Fer"
car2 = Vehicle()
car2.color = "blue"
car2.kind = "van"
car2.value = 10000.00
car2.name = "Jump"
# test code
print(car1.description())
print(car2.description())
```
