```python
sismos = open("signif.txt")
```

```python
sis = []
for linea in sismos:
    sis.append(linea)
sismos.close()   
```

```python
nombres = sis[0].split()
long_id = nombres.index("LONGITUDE")
lat_id = nombres.index("LATITUDE")
longitud = []
latitud = []  
```

```python
lineas_invalidas_long = []
lineas_invalidas_lat = []
i = 0;
for linea in sis:    
    linea_sep = linea.split('\t')
    try:
        float(linea_sep[long_id])                 
    except:
        lineas_invalidas_long.append(i)        
    try:
        float(linea_sep[lat_id])                 
    except:
        lineas_invalidas_lat.append(i)        
    i = i+1
```

```python
print(lineas_invalidas_long)
```

```python
print(lineas_invalidas_lat)
```

```python
lineas_invalidas = list(set(lineas_invalidas_long) | set(lineas_invalidas_lat))
print(lineas_invalidas_lat)
```

```python
i = 0
for linea in sis:
    linea_sep = linea.split('\t')
    if(not(i in lineas_invalidas)):    
        longitud.append(float(linea_sep[long_id]))
        latitud.append(float(linea_sep[lat_id]))
    i = i+1
```
